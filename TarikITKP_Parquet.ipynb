```python
# Fungsi Download JSON
def readJSONLKPP(jenisdata, kode, tahun):
  # Kode berdasarkan lokasi UKPBJ
  if kode == "PROV":
    kodeRUP = "D197"
    kodeLPSE = "97"
  if kode == "SKW":
    kodeRUP = "D200"
    kodeLPSE = "132"
  if kode == "PTK":
    kodeRUP = "D199"
    kodeLPSE = "62"
  if kode == "SBS":
    kodeRUP = "D208"
    kodeLPSE = "354"
  if kode == "MPW":
    kodeRUP = "D552"
    kodeLPSE = "118"
  if kode == "KKR":
    kodeRUP = "D202"
    kodeLPSE = "188"
  if kode == "BKY":
    kodeRUP = "D206"
    kodeLPSE = "444"
  if kode == "LDK":
    kodeRUP = "D205"
    kodeLPSE = "496"  
  if kode == "SGU":
    kodeRUP = "D204"
    kodeLPSE = "298"
  if kode == "SKD":
    kodeRUP = "D198"
    kodeLPSE = "175"
  if kode == "STG":
    kodeRUP = "D211"
    kodeLPSE = "345"
  if kode == "MLW":
    kodeRUP = "D210"
    kodeLPSE = "540"
  if kode == "KPH":
    kodeRUP = "D209"
    kodeLPSE = "488"
  if kode == "KTP":
    kodeRUP = "D201"
    kodeLPSE = "110"
  if kode == "KKU":
    kodeRUP = "D207"
    kodeLPSE = "393"

  # Membuat parameter service data LKPP
  ## Data sipraja
  if jenisdata == "sipraja":
    kodeJSON = f"{tahun}"
    urlFile = f"{urlSIPRAJA}{kodeJSON}"
  ## Data sanggah
  if jenisdata == "sanggah":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlSANGGAH}{kodeJSON}"
  ## Data sirup paket penyedia
  if jenisdata == "sirupdp":
    kodeJSON = f"{tahun}:{kodeRUP}"
    urlFile = f"{urlPP}{kodeJSON}"
  ## Data sirup paket anggaran penyedia
  if jenisdata == "siruppap":
    kodeJSON = f"{tahun}:{kodeRUP}"
    urlFile = f"{urlPAP}{kodeJSON}"
  ## Data sirup swakelola
  if jenisdata == "sirupdsw":
    kodeJSON = f"{tahun}:{kodeRUP}"
    urlFile = f"{urlPS}{kodeJSON}"
  ## Data pencatatan non tender
  if jenisdata == "pntender":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlPNT}{kodeJSON}"
  ## Data pencatatan swakelola
  if jenisdata == "pswakelola":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlPSW}{kodeJSON}"
  ## Data sirup struktur anggaran pemda
  if jenisdata == "sirupdsa_rsap":
    kodeJSON = f"{tahun}:{kodeRUP}"
    urlFile = f"{urlRSAP}{kodeJSON}"
  ## Data sirup program master rup
  if jenisdata == "sirupdsa_pmr":
    kodeJSON = f"{tahun}:{kodeRUP}"
    urlFile = f"{urlPMR}{kodeJSON}"
  ## Data sirup kegiatan master rup
  if jenisdata == "sirupdsa_kmr":
    kodeJSON = f"{tahun}:{kodeRUP}"
    urlFile = f"{urlKMR}{kodeJSON}"
  ## Data sirup rinci objek akun master rup
  if jenisdata == "sirupdsa_roamr":
    kodeJSON = f"{tahun}:{kodeRUP}"
    urlFile = f"{urlROAMR}{kodeJSON}"
  ## Data sirup objek akun master rup
  if jenisdata == "sirupdsa_oamr":
    kodeJSON = f"{tahun}:{kodeRUP}"
    urlFile = f"{urlOAMR}{kodeJSON}"
  ## Data tender pengumuman detail LPSE
  if jenisdata == "dtender_dtp":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDTP}{kodeJSON}"
  ## Data tender selesai
  if jenisdata == "dtender_dts":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDTS}{kodeJSON}"
  ## Data tender kontrak sppbj
  if jenisdata == "dtender_dtks":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDTKS}{kodeJSON}"
  ## Data tender kontrak bap
  if jenisdata == "dtender_dtkb":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDTKB}{kodeJSON}"
  ## Data tender pokja
  if jenisdata == "dtender_dtpo":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDTPO}{kodeJSON}"
  ## Data tender ppk
  if jenisdata == "dtender_dtppk":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDTPPK}{kodeJSON}"
  ## Data non tender e-kontrak sppbj
  if jenisdata == "dntender_dntks":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDNTKS}{kodeJSON}"
  ## Data non tender selesai
  if jenisdata == "dntender_dnts":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDNTS}{kodeJSON}"
  ## Data non tender pengumuman
  if jenisdata == "dntender_dntp":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDNTP}{kodeJSON}"
  ## Data non tender kontrak bap
  if jenisdata == "dntender_dntkb":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDNTKB}{kodeJSON}"
  ## Data non tender pp
  if jenisdata == "dntender_dntpp":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDNTPP}{kodeJSON}"
  ## Data non tender pokja
  if jenisdata == "dntender_dntpo":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDNTPO}{kodeJSON}"
  ## Data non tender ppk
  if jenisdata == "dntender_dntppk":
    kodeJSON = f"{tahun}:{kodeLPSE}"
    urlFile = f"{urlDNTPPK}{kodeJSON}"
  ## Data e-katalog informasi utama paket kldi
  if jenisdata == "dkatalog_deiupk":
    kodeJSON = f"{tahun}:{kodeRUP}"
    urlFile = f"{urlDEIUPK}{kodeJSON}"

  fileCSV = f"{direktori_itkp}{kode}/{jenisdata}{tahun}.csv.gz"
  fileCSVBulan = f"{direktori_itkp}{kode}/{jenisdata}{tahun}{str(x.month)}.csv.gz"
  filePARQUET = f"{direktori_itkp}{kode}/{jenisdata}{tahun}.parquet"
  filePARQUETBulan = f"{direktori_itkp}{kode}/{jenisdata}{tahun}{str(x.month)}.parquet"

  # Cek tahun service yang mau dibaca
  if tahun == "2023":  # nanti kalo udah masuk tahun real time ganti -> tahun == str(x.year)
    # Jika tahun = tahun real time, langsung baca dan konversi service JSON
    try:
      bacaJSON = pd.read_json(urlFile)
      print(f"Baca service data {kode} - {jenisdata} tahun {tahun} berhasil.")
    except:
      print(f"Gagal saat membaca service data JSON LKPP di {urlFile}")
    try:
      bacaJSON.to_csv(fileCSV, compression="gzip")
      bacaJSON.to_csv(fileCSVBulan, compression="gzip")
      print(f"Konversi service data {kode} - {jenisdata} tahun {tahun} ke format CSV berhasil.")
    except:
      print(f"Gagal saat konversi service data ke format CSV {fileCSV}")
    try:
      bacaJSON.to_parquet(filePARQUET)
      bacaJSON.to_parquet(filePARQUETBulan)
      print(f"Konversi service data {kode} - {jenisdata} tahun {tahun} ke format PARQUET berhasil.")
    except:
      print(f"Gagal saat konversi service data ke format PARQUET {filePARQUET}")
  else:
    # Jika tahun != tahun realtime, cek file dulu, kalau sudah ada jangan baca lagi JSON LKPP
    if str(x.day) == "15":
      print(f"Tanggal 15, data CSV dan PARQUET lama akan diperbarui.")
      try:
        bacaJSON = pd.read_json(urlFile)
        print(f"Baca service data {kode} - {jenisdata} tahun {tahun} berhasil.")
      except:
        print(f"Gagal saat membaca service data JSON LKPP di {urlFile}")
      try:
        bacaJSON.to_csv(fileCSV, compression="gzip")
        print(f"Konversi service data {kode} - {jenisdata} tahun {tahun} ke format CSV berhasil.")
      except:
        print(f"Gagal saat konversi service data ke format CSV {fileCSV}")
      try:
        bacaJSON.to_parquet(filePARQUET)
        print(f"Konversi service data {kode} - {jenisdata} tahun {tahun} ke format PARQUET berhasil.")
      except:
        print(f"Gagal saat konversi service data ke format PARQUET {filePARQUET}")
    else:
      if (os.path.exists(fileCSV) or os.path.exists(filePARQUET)):
        print(f"Data {jenisdata} dalam format CSV dan PARQUET tahun {tahun} sudah ada, tidak perlu baca ulang.")
      else:
        try:
          bacaJSON = pd.read_json(urlFile)
          print(f"Baca service data {kode} - {jenisdata} tahun {tahun} berhasil.")
        except:
          print(f"Gagal saat membaca service data JSON LKPP di {urlFile}")
        try:
          bacaJSON.to_csv(fileCSV, compression="gzip")
          print(f"Konversi service data {kode} - {jenisdata} tahun {tahun} ke format CSV berhasil.")
        except:
          print(f"Gagal saat konversi service data ke format CSV {fileCSV}")
        try:
          bacaJSON.to_parquet(filePARQUET)
          print(f"Konversi service data {kode} - {jenisdata} tahun {tahun} ke format PARQUET berhasil.")
        except:
          print(f"Gagal saat konversi service data ke format PARQUET {filePARQUET}")
```

```python
# Looping baca, konversi dan gabung service data JSON LKPP
kodes = ["PROV", "PTK", "KKR", "MPW", "SKW", "SBS", "BKY", "LDK", "SGU", "SKD", "STG", "MLW", "KPH", "KKU", "KTP"]
jenisdatas = ["sipraja", "sanggah", "sirupdp", "siruppap", "sirupdsw", "pntender", "pswakelola", "sirupdsa_rsap", 
              "sirupdsa_pmr", "sirupdsa_kmr", "sirupdsa_roamr", "sirupdsa_oamr", "dtender_dtp", "dtender_dts", 
              "dtender_dtks", "dtender_dtkb", "dtender_dtpo", "dtender_dtppk", "dntender_dntks", "dntender_dnts", 
              "dntender_dntp", "dntender_dntkb", "dntender_dntpp", "dntender_dntpo", "dntender_dntppk"]
tahuns = ["2020", "2021", "2022", "2023"]

# Baca dan konversi service data JSON LKPP
for kode in kodes:
  for jenisdata in jenisdatas:
    for tahun in tahuns:
      readJSONLKPP(jenisdata, kode, tahun)
```
