```python
# SETTING DIREKTORI & URL API JSON LKPP #
# Sampel: https://inaproc.lkpp.go.id/isb/api/35a90e23-7000-4914-a53b-70ba872dc142/json/180003027/PaketPenyediaOpt1618/tipe/4:12/parameter/2021:D197 

## Kode RUP dan Kode LPSE ##
### Silahkan disesuaikan dengan kode RUP dan kode LPSE dari UKPBJ anda ###
#kodeRUP = "D197"
#kodeLPSE = "97"

## Setting Direktori tempat menyimpan file di Google Drive ##
### Silahkan disesuaikan dengan settingan Google Drive di PC anda ###
direktorijson = "/content/drive/MyDrive/Coding/Python/Project/dash-ukpbj/data/json/"
direktoricsv = "/content/drive/MyDrive/Coding/Python/Project/dash-ukpbj/data/csv/"
direktorixlsx = "/content/drive/MyDrive/Coding/Python/Project/dash-ukpbj/data/excel/"
direktoriftr = "/content/drive/MyDrive/Coding/Python/Project/dash-ukpbj/data/ftr/"

direktorigabung = "/content/drive/MyDrive/Coding/Python/Project/dash-ukpbj/data/"

## Data eKatalog
### 1 - Data Katalog - Dibiarkan saja
urlTRXKAT = "https://isb.lkpp.go.id/isb/api/c3dee072-3ef1-4b24-8b1d-7239e1ccd7e9/json/736987130/Ecat-PaketEPurchasing/tipe/4:12/parameter/"
urlTRXKATAYANG = "https://isb.lkpp.go.id/isb/api/4bb5a044-79db-4041-9545-acc042ddc1b2/json/736990830/Ecat-ProdukTayangKatalogLokalSektoral/tipe/12/parameter/"
urlTRXKATKOM = "https://isb.lkpp.go.id/isb/api/cd5fecd9-2bab-4b43-8687-98c5f6604319/json/736987127/Ecat-KomoditasDetail/tipe/4/parameter/"

## Data Toko Daring
urlDARING = "https://inaproc.lkpp.go.id/isb/api/4eb548e7-39cc-4ec8-a4db-8c4d702e0e02/json/736989901/TokoDaringRealisasi/tipe/12:4/parameter/"
```

```python
# Fungsi untuk download data JSON
def unduhJSONTRXKATALOG(jenisdata,kode,tahun):

  if kode == "PROV":
    kodeRUP = "D197"
    kodeLPSE = "97"
  if kode == "SKW":
    kodeRUP = "D200"
    kodeLPSE = "132"
  if kode == "PTK":
    kodeRUP = "D199"
    kodeLPSE = "62"
  if kode == "SBS":
    kodeRUP = "D208"
    kodeLPSE = "354"
  if kode == "MPW":
    kodeRUP = "D552"
    kodeLPSE = "118"
  if kode == "KKR":
    kodeRUP = "D202"
    kodeLPSE = "188"
  if kode == "BKY":
    kodeRUP = "D206"
    kodeLPSE = "444"
  if kode == "LDK":
    kodeRUP = "D205"
    kodeLPSE = "496"  
  if kode == "SGU":
    kodeRUP = "D204"
    kodeLPSE = "298"
  if kode == "SKD":
    kodeRUP = "D198"
    kodeLPSE = "175"
  if kode == "STG":
    kodeRUP = "D211"
    kodeLPSE = "345"
  if kode == "MLW":
    kodeRUP = "D210"
    kodeLPSE = "540"
  if kode == "KPH":
    kodeRUP = "D209"
    kodeLPSE = "488"
  if kode == "KTP":
    kodeRUP = "D201"
    kodeLPSE = "110"
  if kode == "KKU":
    kodeRUP = "D207"
    kodeLPSE = "393"
  if kode == "KALSEL":
    kodeRUP = "D212"
    kodeLPSE = "181"
  if kode == "BPK":
    kodeRUP = "L13"

  if jenisdata == "trxkatalog":
    # Download Data Transaksi Katalog
    kodeJSON = tahun + ":" + kodeRUP
    # Variabel url Data Transaksi Katalog
    urlfile = urlTRXKAT + kodeJSON

  if jenisdata == "prodkatalog":
    # Download Data Produk Katalog Tayang
    kodeJSON = kodeRUP
    # Variabel url Data Produk Tayang
    urlfile = urlTRXKATAYANG + kodeJSON

  if jenisdata == "daring":
    # Download Data Transaksi Toko Daring
    kodeJSON = kodeRUP + ":" + tahun
    # Variabel url Toko Daring
    urlfile = urlDARING + kodeJSON

  fileJSON = direktorijson + kode + "/" + jenisdata + tahun + ".json"
  fileCSV = direktoricsv + kode + "/" + jenisdata + tahun + ".csv"
  fileXLSX = direktorixlsx + kode + "/" + jenisdata + tahun + ".xlsx"
  fileFTR = direktoriftr + kode + "/" + jenisdata + tahun + ".ftr"

  ## Download dana Konversi Data ##
  if tahun != "2022":
    if (os.path.exists(fileJSON) or os.path.exists(fileCSV) or os.path.exists(fileXLSX) or os.path.exists(fileFTR)):
      print("Data JSON, CSV, EXCEL dan FTR tahun " + tahun + " sudah ada, tidak perlu di download lagi !")
    else:
      if (wget.download(urlfile,fileJSON)):
        print("Download data " + kode + " - " + jenisdata + " Tahun " + tahun + " Berhasil.")
        # Konversi data JSON to CSV, EXCEL, dan FTR
        try:
          dfUnduh = pd.read_json(fileJSON)
        except:
          print("Terjadi masalah saat membaca file JSON ", fileJSON)
        try:
          dfUnduh.to_csv(fileCSV)
        except:
          print("Terjadi masalah saat konversi file JSON ke CSV ", fileCSV)
        try:
          dfUnduh.to_excel(fileXLSX, index = False)
        except:
          print("Terjadi masalah saat konversi file JSON ke EXCEL ", fileXLSX)
        try:
          dfUnduh.to_feather(fileFTR)
        except:
          print("Terjadi masalah saat konversi file JSON ke Feather ", fileFTR)
      else:
        print("Download data " + kode + " - " + jenisdata + " Tahun " + tahun + " Gagal.")        
  else:
    if (os.path.exists(fileJSON) or os.path.exists(fileCSV) or os.path.exists(fileXLSX) or os.path.exists(fileFTR)):
      try:
        os.remove(fileJSON)
      except:
        print("Terjadi masalah saat hapus file ", fileJSON)
      try:
        os.remove(fileCSV)
      except:
        print("Terjadi masalah saat hapus file ", fileCSV)
      try:
        os.remove(fileXLSX)
      except:
        print("Terjadi masalah saat hapus file ", fileXLSX)
      try:
        os.remove(fileFTR)
      except:
        print("Terjadi masalah saat hapus file ", fileFTR)

      if (wget.download(urlfile,fileJSON)):
        print("Sukses Update Data JSON, CSV, EXCEL dan FTR - " + kode + " - " + jenisdata + " Tahun " + tahun + ".")
        # Konversi data JSON to CSV, EXCEL, dan FTR
        try:
          dfUnduh = pd.read_json(fileJSON)
        except:
          print("Terjadi masalah saat membaca file JSON ", fileJSON)
        try:
          dfUnduh.to_csv(fileCSV)
        except:
          print("Terjadi masalah saat konversi file JSON ke CSV ", fileCSV)
        try:
          dfUnduh.to_excel(fileXLSX, index = False)
        except:
          print("Terjadi masalah saat konversi file JSON ke EXCEL ", fileXLSX)
        try:
          dfUnduh.to_feather(fileFTR)
        except:
          print("Terjadi masalah saat konversi file JSON ke Feather ", fileFTR)
      else:
        print("Gagal Update Data JSON, CSV, EXCEL dan FTR - " + kode + " - " + jenisdata + " Tahun " + tahun + ".")
    else:
      if (wget.download(urlfile,fileJSON)):
        print("Download data " + kode + " - " + jenisdata + " Tahun " + tahun + " Berhasil.")
        # Konversi data JSON to CSV, EXCEL, dan FTR
        try:
          dfUnduh = pd.read_json(fileJSON)
        except:
          print("Terjadi masalah saat membaca file JSON ", fileJSON)
        try:
          dfUnduh.to_csv(fileCSV)
        except:
          print("Terjadi masalah saat konversi file JSON ke CSV ", fileCSV)
        try:
          dfUnduh.to_excel(fileXLSX, index = False)
        except:
          print("Terjadi masalah saat konversi file JSON ke EXCEL ", fileXLSX)
        try:
          dfUnduh.to_feather(fileFTR)
        except:
          print("Terjadi masalah saat konversi file JSON ke Feather ", fileFTR)
      else:
        print("Download data " + kode + " - " + jenisdata + " Tahun " + tahun + " Gagal.")
```

```python
pilihs = ["tarikdatakatalog", "gabungdatakatalog", "tarikdatakomoditas", "katalogdetail"]
for pilih in pilihs:

  if pilih == "tarikdatakatalog":

    # Unduh Data E-Purchasing
    kldis = ["PROV", "PTK", "KKR", "MPW", "SKW", "SBS", "BKY", "LDK", "SGU", "SKD", "STG", "MLW", "KPH", "KKU", "KTP", "BPK"]
    jenisdatas = ["trxkatalog", "prodkatalog", "daring"]
    tahuns = ["2019","2020","2021","2022"]

    #kldi = input("KLDI ? (Format: PROV, PTK, KKR ...")
    #kode = kldi.upper()

    for kode in kldis:
      for jenisdata in jenisdatas:
        for tahun in tahuns:
          unduhJSONTRXKATALOG(jenisdata,kode,tahun)

  elif pilih == "gabungdatakatalog":

    # Gabung Data Transaksi E-Katalog dan Toko Daring
    provs = ["PROV", "PTK", "KKR", "MPW", "SKW", "SBS", "BKY", "LDK", "SGU", "SKD", "STG", "MLW", "KPH", "KKU", "KTP", "BPK"]

    trxlist = []
    prodlist = []
    tdlist = []

    for prov in provs:
      # Gabung Data Transaksi Katalog
      trxFTR = direktoriftr + prov + "/trxkatalog2022.ftr"
      if (os.path.exists(trxFTR)):
        trxkat = pd.read_feather(trxFTR)       
        trxlist.append(trxkat)
      # Gabung Data Produk Katalog
      prodFTR = direktoriftr + prov + "/prodkatalog2022.ftr"
      if (os.path.exists(prodFTR)):
        prodkat = pd.read_feather(prodFTR)
        prodlist.append(prodkat)
      # Gabung Data Transaksi Toko Daring
      daringFTR = direktoriftr + prov + "/daring2022.ftr"
      if (os.path.exists(daringFTR)):
        trxtd = pd.read_feather(daringFTR)
        tdlist.append(trxtd)
  
    fileKATGABXLSX = direktorigabung + "trxkatalog.xlsx"
    filePRODGABXLSX = direktorigabung + "prodkatalog.xlsx"
    fileDARINGGABXLSX = direktorigabung + "daring.xlsx"

    ## Gabung Data Transaksi E-Katalog
    try:
      trxgabung = pd.concat(trxlist, ignore_index = True)
      print("Penggabungan data E-Katalog Kalbar Sukses")   
    except:
      print("Terjadi masalah saat menggabungkan data E-Katalog Kalbar")
    try:
      trxgabung.to_excel(fileKATGABXLSX, index = False)
      print("Konversi Data E-Katalog Kalbar ke format Excel Sukses")
    except:
      print("Terjadi masalah saat konversi Data E-Katalog Kalbar ke format Excel")

    ## Gabung Data Produk E-Katalog
    try:
      prodgabung = pd.concat(prodlist, ignore_index = True)
      print("Penggabungan data Produk E-Katalog Kalbar Sukses")   
    except:
      print("Terjadi masalah saat menggabungkan data Produk E-Katalog Kalbar")
    try:
      prodgabung.to_excel(filePRODGABXLSX, index = False)
      print("Konversi Data Produk E-Katalog Kalbar ke format Excel Sukses")
    except:
      print("Terjadi masalah saat konversi Data Produk E-Katalog Kalbar ke format Excel")

    ## Gabung Data Transaksi Toko Daring
    try:
      daringgabung = pd.concat(tdlist, ignore_index = True)
      print("Penggabungan data Transaksi Toko Daring Kalbar Sukses")   
    except:
      print("Terjadi masalah saat menggabungkan data Transaksi Toko Daring Kalbar")
    try:
      daringgabung.to_excel(fileDARINGGABXLSX, index = False)
      print("Konversi data Transaksi Toko Daring Kalbar ke format Excel Sukses")
    except:
      print("Terjadi masalah saat konversi data Transaksi Toko Daring Kalbar ke format Excel")

  elif pilih == "tarikdatakomoditas":

    # Tarik Data Komoditas
    trxkataloggabung = direktorigabung + "trxkatalog.xlsx"
    trxkatalog = pd.read_excel(trxkataloggabung)
    
    fileKOMDETXLSX = direktorigabung + "komoditasdetail.xlsx"
    
    kodekomoditas = trxkatalog['kd_komoditas'].unique()
    komoditasgab = []
    for kodekom in kodekomoditas:
      urlkomoditi = urlTRXKATKOM + str(kodekom)
      komoditas = pd.read_json(urlkomoditi)
      komoditasgab.append(komoditas)

    try:
      komoditasdetail = pd.concat(komoditasgab, ignore_index = True)
      print("Penggabungan data Komoditas Detail Kalbar Sukses")
    except:
      print("Terjadi masalah saat menggabungkan data Komoditas Detail Kalbar")
    try:
      komoditasdetail.to_excel(fileKOMDETXLSX, index = False)
      print("Konversi Data Komoditas Detail Kalbar ke format Excel Sukses")
    except:
      print("Terjadi masalah saat konversi Data Komoditas Detail Kalbar ke format Excel")

  elif pilih == "katalogdetail":

    # Tarik Data Transaksi
    trxkataloggabung = direktorigabung + "trxkatalog.xlsx"
    trxkatalog = pd.read_excel(trxkataloggabung)
    # Tarik Data Detail Komoditas
    komoditasdetailgabung = direktorigabung + "komoditasdetail.xlsx"
    komoditasdetail = pd.read_excel(komoditasdetailgabung)

    fileKATALOGDETAILXLSX = direktorigabung + "katalogdetail.xlsx"

    try:
      katalog_detail = trxkatalog.merge(komoditasdetail, how='left', on='kd_komoditas')
      print("Penggabungan data Transaksi dan Detail Komoditas Katalog Sukses")
    except:
      print("Terjadi masalah saat menggabungkan data Katalog Detail Kalbar")

    try:
      katalog_detail.to_excel(fileKATALOGDETAILXLSX, index = False)
      print("Konversi Data Katalog Detail Kalbar ke format Excel Sukses")
    except:
      print("Terjadi masalah saat konversi data Katalog Detail Kalbar")

  else:
    print("Perintah yang anda inginkan tidak tersedia ... !")  
```
